- A linear regression algorithm is a ML algorithm based on supervised learning.

- It performs a regression task on the basis of two variables (independent & dependent)

- It is mostly used for forecasting & finding out relationship between two variables


**Equation**

y = b0+b1x1+e

where y ---> dependent variable

      x ---> independent variable
      
      b ---> coefficient 
      
      e ---> error (average is 0)

Price -- independent    

Phone Model -- dependent 



Area -- independent

Price -- dependent 


Loan Amt -- independent

Salary -- dependent

## Prediction 

y = b0+b1*x1

b0 ==> 0.2750

b1 ==> 0.0017

```python
b0 = 0.2750
b1 = 0.0017


def gpa_predictor(score):
    y = b0+b1*score
    return y
```

```python
from sklearn.linear_model import LinearRegression
```

```python
pred = lr.predict(x_test)
```

```python
len(pred)
```

```python
print('Training Accuracy is',lr.score(x_train,y_train))
```

```python
print('Testing Accuracy is',lr.score(x_test,y_test))
```
